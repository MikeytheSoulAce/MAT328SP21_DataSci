```python
import pandas as pd
import seaborn as sns
```

```python
covid_data = pd.read_csv("data-by-modzcta.csv")
```

```python
# standard pairplot with relevant quantitative columns
sns.pairplot(data=covid_data, vars=["COVID_CASE_COUNT", "COVID_CASE_RATE", "COVID_DEATH_COUNT", "COVID_DEATH_RATE", "PERCENT_POSITIVE", "TOTAL_COVID_TESTS"], hue="BOROUGH_GROUP")
```

```python
# Interesting graph that depicts lat/long of ZCTAs
sns.pairplot(data=covid_data, vars=["lat", "lon"], hue="BOROUGH_GROUP")
```

```python
# Heatmap of all quantitative columns
sns.heatmap(covid_data.corr())
```

```python
# using PairGrid to draw 
sns.set_theme(style="white")

g = sns.PairGrid(covid_data, vars=["COVID_CASE_COUNT", "COVID_CASE_RATE", "COVID_DEATH_COUNT", "COVID_DEATH_RATE", "PERCENT_POSITIVE", "TOTAL_COVID_TESTS"], diag_sharey=False)
g.map_upper(sns.scatterplot, s=15)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=2)
```
